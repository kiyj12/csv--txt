```python
# 한 번에 놓음.

import numpy as np
import pandas as pd
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)

import warnings
warnings.filterwarnings(action='ignore')

df = pd.DataFrame(columns=['제목','가수', '가사'])

from os import listdir
from os.path import isfile, join
files = [f for f in listdir('C:/Users/kiyj1/OneDrive - 연세대학교 (Yonsei University)/문서/LISL/멜론 가사/여돌 가사 txt 파일/곡별 여돌 가사/아이돌 가사 2014 텍파-강경') if isfile(join('C:/Users/kiyj1/OneDrive - 연세대학교 (Yonsei University)/문서/LISL/멜론 가사/여돌 가사 txt 파일/곡별 여돌 가사/아이돌 가사 2014 텍파-강경', f))]

df2 = df.append(files)

df2.columns=['제목', '가수', '가사','0']

df2.drop('0', axis=1, inplace=True)

#데이터프레임에서 문자열 분리, 특정 열에 넣기

df2['가수'] = df2['제목'].str.split("-").str[1]

# 데이터프레임 안에서 문자열 편집

df2['제목'] = df2['제목'].str.replace(' - .+txt','')
df2['가수'] = df2['가수'].str.replace('_.+txt','')


df = df2

# 가사 텍파들 읽어와서 데이터프레임에 넣기
import os

basedir='C:/Users/kiyj1/OneDrive - 연세대학교 (Yonsei University)/문서/LISL/멜론 가사/여돌 가사 txt 파일/곡별 여돌 가사/아이돌 가사 2014 텍파-강경/'

llist = []

files = os.listdir(basedir)

for filename in files:
    if ".txt" not in filename:
        continue
    try:
        file = open(basedir + filename, 'rt', encoding='utf-8')
        contents = file.readlines()
        llist.append(contents)
    except:
        file = open(basedir + filename, 'rt', encoding='euc-kr')
        contents = file.readlines()
        llist.append(contents)

alist=[]

for i in llist:
    alist.append(''.join(i))
    

data = pd.DataFrame()

data = data.append(alist)

df2 = pd.concat([df2, data], axis = 1)
df2 = df2.replace(r'\n', '', regex=True)
del df2['가사']

df2.rename(columns={0:'가사'}, inplace=True)

df2.to_excel('[강경]2014 여돌 가사.xlsx', index=False)
```
